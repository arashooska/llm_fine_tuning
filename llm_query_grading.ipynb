#### Grading Rubrics

<b>Content Rubric:
- 1 pt for being on topic (even remotely about the question)
- 1 pt if the answer is >=50% coherent
- 1 pt if the answer is fully coherent (no words out of place etc)
- 1 pt for having a reasonable answer 
- 1 extra point for a thorough, non-trivial answer

<b>Grammar Rubric:
- 1 pt for all words being real words
- 1 pt for reasonable amount of English
- 1 pt for subject verb agreement
- 1 pt for correct word order
- 1 pt for proper punctuation
