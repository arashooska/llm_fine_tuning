```python
TRAIN_FP = "/raid/aransari/datasets/PCoQA/Data/PCoQA_Train.pk"
```

```python
import pickle

with open(TEST_FP, "rb") as f:
    train_data = pickle.load(f)
```

```python
# Now we write a function to separate out every 'question' and it's corresponding 'human_answer' into the following format:
# [{input: 'question', output: 'human_answer'}, {...}, ...]

def extract_qa(data):
    qa_pairs = []
    for entry in data:
        for qa in entry['qas']:
            qa_pairs.append({'input': qa['question'], 'output': qa['human_answer'][0]['text']})
    return qa_pairs



```

```python

qa_pairs = extract_qa(train_data)
```

```python
qa_pairs
```

```python
field_names = ['input', 'output']
```

```python
#Export the test data to csv
def export_dict_to_csv(dict, field_names, fpath):
    import csv
    with open(fpath, 'w') as csvfile: 
        writer = csv.DictWriter(csvfile, fieldnames = field_names) 
        writer.writeheader() 
        writer.writerows(dict) 
```

```python
export_dict_to_csv(qa_pairs, field_names, "/raid/aransari/datasets/pcoqa_test_data_clean.csv")
```

```python
#Do the same thing for train data

import pickle

with open(TRAIN_FP, "rb") as f:
    train_data = pickle.load(f)

qa_pairs_train = extract_qa(train_data)

export_dict_to_csv(qa_pairs_train, field_names, "/raid/aransari/datasets/pcoqa_train_data_clean.csv")
```
